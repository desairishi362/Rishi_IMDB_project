# 14. Find Most Popular Movie Title (Higest Revenue)

```python

highest_revenue_movie = data.loc[data['Revenue (Millions)'].idxmax(), ['Title', 'Revenue (Millions)']]
print("Most Popular Movie (Highest Revenue):")
print(highest_revenue_movie)



top10 = data.nlargest(10, 'Revenue (Millions)')[['Title', 'Revenue (Millions)']]

plt.figure(figsize=(12,6))
plt.bar(top10['Title'], top10['Revenue (Millions)'], color='skyblue')
plt.title('Top 10 Movies by Revenue', fontsize=14)
plt.xlabel('Movie Title')
plt.ylabel('Revenue (Millions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### From the above analysis the most popular movie title with highest revenue is Star Wars with the revenue of above 800 (Millions) 

# 15. Display Top 10 Highest Rated Movie Titles And its Directors


```python

top10_rated = data.sort_values(by='Rating', ascending=False)[['Title', 'Director']].head(10)

print("Top 10 Highest Rated Movie Titles and Their Directors:")
print(top10_rated)



plt.figure(figsize=(12,6))
plt.barh(top10_rated['Title'], data.sort_values(by='Rating', ascending=False)['Rating'].head(10), color='salmon')
plt.xlabel('Rating')
plt.ylabel('Movie Title')
plt.title('Top 10 Highest Rated Movies')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

### As we can see from the above graph the highest rated movie titles is The Dark Knight

# 16. Display Top 10 Highest Revenue Movie Titles


```python
import pandas as pd

data = pd.read_csv('5-IMDB-Movie-Data.csv')

top10_revenue = data.sort_values(by='Revenue (Millions)', ascending=False)[['Title', 'Revenue (Millions)']].head(10)

print("Top 10 Highest Revenue Movie Titles:")
print(top10_revenue)


plt.figure(figsize=(12,6))
plt.bar(top10_revenue['Title'], top10_revenue['Revenue (Millions)'], color='skyblue')
plt.title('Top 10 Highest Revenue Movies', fontsize=14)
plt.xlabel('Movie Title')
plt.ylabel('Revenue (Millions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```
